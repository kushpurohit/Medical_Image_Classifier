```python
import os
import json
import shutil

def organize_images_by_label(json_file, image_dir, output_dir):
    # Define desired labels
    desired_labels = ["Macroscopy", "BG", "FACS", "Microscopy"]

    # Load the JSON data
    with open(json_file, "r") as f:
        data = json.load(f)

    # Create output folders (if they don't exist)
    for label in desired_labels:
        os.makedirs(os.path.join(output_dir, label), exist_ok=True)

    counter = 1  # Initialize the counter for unique identifiers

    # Process each image based on its label
    for subfolder in os.listdir(image_dir):
        subfolder_path = os.path.join(image_dir, subfolder)
        for filename in os.listdir(subfolder_path):
            image_path = os.path.join(subfolder_path, filename)
            image_name, extension = os.path.splitext(filename)  # Separate filename and extension correctly
            image_name += ".png"
            # Handle potential KeyErrors for missing images in JSON data
            try:
#             if image_name in data[subfolder]:
                label = data[str(subfolder)][image_name]
            except KeyError:
                print(f"Warning: Image name '{image_name}' not found in JSON data.")


            if label in desired_labels:
                # Generate a unique identifier
                unique_id = counter

                # Construct the new path with the unique identifier and extension
                new_path = os.path.join(output_dir, label, f"{image_name}_{unique_id}{extension}")
                os.makedirs(os.path.dirname(new_path), exist_ok=True)  # Ensure parent directories exist
                shutil.move(image_path, new_path)
            else:
                print(f"Warning: Label '{label}' for image '{image_name}' not found in desired labels.")

            counter += 1  # Increment the counter for the next image

# Example usage
json_file = r"C:\Users\Kush\BioFors-main\BioFors-main\annotation_files\classification.json"
image_dir = r"C:\Users\Kush\Scientific_Image_Classifier\data\biofors_images"
output_dir = r"C:\Users\Kush\Scientific_Image_Classifier\data"

organize_images_by_label(json_file, image_dir, output_dir)

print("Image organization completed.")
```
